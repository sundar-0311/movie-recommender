we have a list of dictionaries, convert to list of string with only the genre name

```python
import ast
def convert(obj):
    gen= []
    for i in ast.literal_eval(obj):
        gen.append(i['name'])
    return gen
```

this doesnt work cause the genre is a string of list of dictionary, thus we converr it using a module called ast and use the function literal_eval

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
import ast
def convert_name(obj):
    gen= []
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            gen.append(i['name'])
            count+=1
        else:
            break
    return gen
```

```python
movies['cast']= movies['cast'].apply(convert_name)
```

```python
movies.iloc[0].crew
```

```python
def fetch_dir(obj):
    lst= []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            lst.append(i['name'])
            break
    return lst
```

```python
movies['crew']= movies['crew'].apply(fetch_dir)
```

# VECTORISATION

for converting tag to vector, we use tf-idf

```python
corpus= new_df['tag'].tolist()
## has list of list for 1 column, all rows
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,   # optional
    stop_words='english' # optional
)

X = tfidf.fit_transform(corpus)
```

```python
X.toarray()
```

now, we need to find distance of each movie with all other movies and sort them on ascending order

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(X)
```

```python
similarity
```

here shape is (4806, 4806) as each vector is getting distance with every other vector

similarity of each movie with itself is 1, so all diagonals will be 1

outside array is for each movie, inside array is for distance of that movie with other movies

we will have a problem when we sort, thr inner array is so that each value is distance is arranged in order 1st mov, 2nd mov etc, if we sort them then the value of index will change, so we use enumerate 

```python
sorted_mov= sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x: x[1])
## sorts based on the 
```

```python
similar_mov= sorted_mov[1:6]
```

```python
similar_mov
```

```python
def distances(movie):
    # Find the movie index
    movie_lower = movie.lower()
    movie_indices = new_df[new_df['title'] == movie_lower].index
    
    if len(movie_indices) == 0:
        print(f"Movie '{movie}' not found in database")
        return
    
    mov_idx = movie_indices[0]
    
    # Get similarity distances for this movie
    dist_list = similarity[mov_idx]
    
    # Sort by similarity score (descending) and get top 6 (including the movie itself)
    sorted_mov = sorted(list(enumerate(dist_list)), reverse=True, key=lambda x: x[1])
    
    # Get top 5 recommendations (skip the first one which is the movie itself)
    similar_movies = sorted_mov[1:6]
    
    print(f"Movies similar to '{movie}':")
    for i in similar_movies:
        print(f"  - {new_df.iloc[i[0]].title} (similarity: {i[1]:.4f})")
```

```python
distances('avatar')
```
